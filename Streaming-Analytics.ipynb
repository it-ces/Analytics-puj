```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.pipeline import make_pipeline
import preprocess
```

```python
# Stander scaler what makes?
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns = iris.feature_names)
df['y'] = iris.target
df = df[df['y'] != 2] # Keep only two targets...
```

```python
X = df.drop(columns='y')
y = df['y']
```

```python
X.describe()
```

```python
def X_standard(df_X):
    # Analize this without copy...
    X = df_X.copy()
    for x in X.columns:
        X[x]  = (X[x] - X[x].mean())/X[x].std()
    return X
```

```python
X_standard(X).describe()
```

```python
################################
# Using Sklearn to standardize #
################################
scaler = StandardScaler()
scaler.fit_transform(X) # keep numpy array 
X_standard = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_standard.describe()
```

```python
clf = LogisticRegression(random_state=0).fit(X_standard, y)
preds = clf.predict_proba(X_standard)
# Log - loss
loss = log_loss(y, preds)
print(loss)
```

```python
pipeline = make_pipeline(StandardScaler(), LogisticRegression(random_state=0))
pipeline.fit(X, y)
y_pred_proba = pipeline.predict_proba(X)
loss = log_loss(y, y_pred_proba)
print(loss)
```

```python
############################################
## Our DataFrame                           #
############################################

# Description:  20 observations, income: (0 nulls), age: (4 nulls),   
#               sex (0 nulls balanced ), country (2 nulls, 5 categories with two unlikely)
#               program(0 nulls, 3 categories), work(1 null),
#               churn(0 nulls),  

def data():
      size = 25
      np.random.seed(0)
      income = np.random.normal(130,22,size)
      age  =  np.random.normal(35, 10, size)
      age[np.random.randint(0,size, 4)] = np.nan
      sex  = np.random.choice(['F', 'M'], size=size, p=[0.5, 0.5])
      country = np.random.choice(['Europe', 'Africa', 'Latin', 'USA', 'other'], size = size, p=[0.45, 0.3, 0.15, 0.05, 0.05 ])
      country = country.astype(object) # to include nans...
      country[np.random.randint(0,size, 8)] = np.nan
      program = np.random.choice(['A', 'B', 'C', 'D', 'E', 'F'], size=size, p=[0.4,0.3,0.05, 0.05,0.1,0.1])
      program = program.astype(object) # to include nans...
      program[np.random.randint(0,size, 2)] = np.nan
      work = np.random.choice(['yes', 'not'], size=size, p=[0.7, 0.3])
      churn = np.random.choice(['Churn','stay'], size = size, p=[0.20, 0.80])
      df = {'income':income, 'age':age, 'sex':sex, 
            'country':country, 'program':program, 
            'work':work, 'churn':churn}
      df = pd.DataFrame(df)
      return df
df = data()
```

```python
def drop_colNaN(df_, min_percent):
    # If the column have lesser of threshold complete data 
    # will be drop of the dataset...
    # threshold = 1,indicates that keep only variables without nans...
    df = df_.copy()
    N  = df.shape[0] 
    keep = [var for var in df.columns if df[var].notnull().sum()/N >= min_percent]
    return df[keep]
```
